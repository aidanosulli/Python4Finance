## Returns a named column data from a CSV

```python
def get_column_from_csv(file, col_name):
    #Try to get the file and if it doesn't exist, issue a warning
    try:
        df = pd.read_csv(file)
    except FileNotFoundError:
        print("File Doesn't Exist")
    else:
        return df[col_name]
```

## Save Stock Data to CSV

```python
def save_to_csv_from_yahoo(folder, ticker):
    stock = yf.Ticker(ticker)
    try:
        print("Get Data for : ", ticker)
        df = stock.history(period = "max")
        time.sleep(2)
        if df.empty:
            stocks_not_downloaded.append(ticker)
        the_file = folder + ticker.replace(".", "_") + '.csv' #can't have any extra periods
        print(time, "Saved")
        df.to_csv(the_file)
    except Exception as ex:
        stocks_not_downloaded.append(ticker)
        print("Couldn't get data for : ", ticker)
```

## Get 5 years of data for the 1st 20 stocks


```python
folder = PATH + '/Wilshire_Stocks/'
for x in range(1001, 2000):
    save_to_csv_from_yahoo(folder, tickers[x])
print("Finished")
```
